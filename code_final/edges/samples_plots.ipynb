```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
csq = pd.read_csv('csq.gz', sep = " ", header= None)
samples= pd.read_csv('samples.gz', sep = " ", header= None)
```

```python
csq = csq.values
samples = samples.values
```

```python
np.mean(samples, axis=0)
```

```python
params_cut = np.copy(samples)
fig1, ax_list = plt.subplots(3, 2, figsize=(10,10))
fig1.suptitle('Corner plots of the samples', fontsize=16)

ax_list[0, 0].plot(params_cut[:, 0], params_cut[:, 1], linestyle = "", marker=".")
ax_list[0, 0].set_ylabel('param 1', fontsize=12)
ax_list[0, 0].set_xlabel('param 0', fontsize=12)

ax_list[0, 1].plot(params_cut[:, 0], params_cut[:, 2], linestyle = "", marker=".")
ax_list[0, 1].set_ylabel('param 2', fontsize=12)
ax_list[0, 1].set_xlabel('param 0', fontsize=12)
    
ax_list[1, 0].plot(params_cut[:, 0], params_cut[:, 3], linestyle = "", marker=".")
ax_list[1, 0].set_ylabel('param 3', fontsize=12)
ax_list[1, 0].set_xlabel('param 0', fontsize=12)

ax_list[1, 1].plot(params_cut[:, 1], params_cut[:, 2], linestyle = "", marker=".")
ax_list[1, 1].set_ylabel('param 2', fontsize=12)
ax_list[1, 1].set_xlabel('param 1', fontsize=12)

ax_list[2, 0].plot(params_cut[:, 1], params_cut[:, 3], linestyle = "", marker=".")
ax_list[2, 0].set_ylabel('param 3', fontsize=12)
ax_list[2, 0].set_xlabel('param 1', fontsize=12)

ax_list[2, 1].plot(params_cut[:, 2], params_cut[:, 3], linestyle = "", marker=".")
ax_list[2, 1].set_ylabel('param 3', fontsize=12)
ax_list[2, 1].set_xlabel('param 2', fontsize=12)
plt.tight_layout()
plt.show()
```

```python
fig2, ax_list = plt.subplots(2, 2, figsize=(10,10))
ax_list[0, 0].hist(samples[:, 0], bins = 50)
ax_list[0, 0].set_ylabel('$pop\_rad\_yield\_0$', fontsize=12)

ax_list[0, 1].hist(samples[:, 1], bins = 50)
ax_list[0, 1].set_ylabel('$pop\_rad\_yield\_2$', fontsize=12)

ax_list[1, 0].hist(samples[:, 2], bins = 50)
ax_list[1, 0].set_ylabel('$f_{esc}$', fontsize=12)

ax_list[1, 1].hist(samples[:, 3], bins = 50)
ax_list[1, 1].set_ylabel('$f_{X}$', fontsize=12)
plt.tight_layout()
plt.savefig('histograms_edges.png')
plt.show()
```

```python
print(np.mean(samples, axis=0))
print(np.std(samples, axis=0))
```

```python
fig3 = plt.figure()
plt.hist(csq - 8006.655671789552, bins=50)
plt.title('Histogram of disparity between the $\chi^2$ values', fontsize=14)
plt.ylabel('$\Delta \chi^2$', fontsize=12)
plt.xlabel('range', fontsize=12)
plt.xlim(right=1.5)
plt.savefig('csq_hist_edges.png')
plt.show()
```

```python
print(np.std(csq-8006.655671789552))
print(np.mean(csq-8006.655671789552))
```

```python
fig4, ax_list = plt.subplots(2, 2, figsize=(12,10))
fig4.suptitle('$\chi^2$ as a Function of Parameters of The Model', fontsize=16)

ax_list[0, 0].scatter(samples[:, 0], csq)
ax_list[0, 0].set_ylabel('$\chi^2$', fontsize=12)
ax_list[0, 0].set_xlabel('$pop\_rad\_yield\_0$', fontsize=12)

ax_list[0, 1].scatter(samples[:, 1], csq)
ax_list[0, 1].set_ylabel('$\chi^2$', fontsize=12)
ax_list[0, 1].set_xlabel('$pop\_rad\_yield\_2$', fontsize=12)

ax_list[1, 0].scatter(samples[:, 2], csq)
ax_list[1, 0].set_ylabel('$\chi^2$', fontsize=12)
ax_list[1, 0].set_xlabel('$f_{esc}$', fontsize=12)

ax_list[1, 1].scatter(samples[:, 3], csq)
ax_list[1, 1].set_ylabel('$\chi^2$', fontsize=12)
ax_list[1, 1].set_xlabel('$f_{X}$', fontsize=12)
plt.tight_layout()
plt.savefig('csq_vs_params_edges.png')
plt.show()
```

```python
fig4, ax_list = plt.subplots(1, 2, figsize=(12,5))
fig4.suptitle('$\chi^2$ as a Function of Parameters of The Model', fontsize=16)

ax_list[0].scatter(samples[:, 0], csq)
ax_list[0].set_ylabel('$\chi^2$', fontsize=12)
ax_list[0].set_xlabel('$pop\_rad\_yield\_0$', fontsize=12)
ax_list[0].set_ylim(top= 8007, bottom= 0.55 + 8006.655671789552)

ax_list[1].scatter(samples[:, 3], csq)
ax_list[1].set_ylabel('$\chi^2$', fontsize=12)
ax_list[1].set_xlabel('$f_{esc}$', fontsize=12)
ax_list[1].set_ylim(top= 8007, bottom= 0.55 + 8006.655671789552)

plt.tight_layout()
plt.savefig('csq_vs_params_zoomed_edges.png')
plt.show()
```
